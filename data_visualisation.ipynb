```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Name":["A","B","C","D","E","R","S","K"],
    "Age":[22,25,24,23,26,27,28,22],
    "Gender":["M","F","M","F","M","F","M","F"],
    "Marks":[75,88,92,70,85,90,78,95]}
plt.plot(data["Marks"]) #univeriate analysis
plt.title("Marks trend")
plt.xlabel("index")
plt.ylabel("Marks")
plt.show()

```

```python
plt.bar(data["Name"],data["Marks"])
plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```

```python
plt.hist(data["Age"],bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(data["Age"],data["Marks"])
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("MArks")
plt.show()
```
