```python
# creating arrays using built in methods
arr3=np.zeros((2,3), dtype="int32")#prefill
print(arr3, arr3.shape)

arr4=np.ones((5,),dtype="int32")
print(arr4, arr4.shape)

arr5=np.full((2,3),20)
print(arr5,arr5.dtype,arr5.shape)

# creationg identity matrix
arr6=np.eye(3,dtype="int32")
print(arr6,type(arr6),arr6.shape,arr6.dtype)

# arr in range with jump by 2
arr7=np.arange(0,11,2)
print(arr7,arr7.shape)

# evenly spaced array
arr8=np.linspace(0,3,6) #linspace(start,end,no of elements)
print(arr8)

```

```python
# properties numpy array
arr1=np.array([[1,2,3],[4,5,6],[1,2,3]])
print(arr1.ndim)
# arr.shape   dimensions output
print(arr1.size)
print(arr1.dtype)
 # no of dimension
print(arr1.ndim)  #2 bcz its 2d array

# typecasting
float_arr=arr1.astype(np.float64)
print(float_arr,float_arr.dtype,float_arr.size,float_arr.ndim)

```

```python
# inportant operation on numpy arrays 
 # 1 2 3 
 # 4 5 6

# to

#  1 2 
#  3 4
#  5 6

arr=np.array([[1,2,3],[4,5,6]])
print(arr, arr.shape)

reshaped=arr.reshape((3,2))
print(reshaped,reshaped.size,reshaped.dtype,reshaped.ndim)

flattened= arr.flatten()         #2d to 1d

print(flattened,flattened.size,flattened.shape,flattened.ndim)
```

```python
# axes in numpy

arr=np.array([[1,2,3],[4,5,6]])
print(np.sum(arr,axis=0))# represents column sum but axis =0 is row vertically
print(np.sum(arr,axis=1))#represents row wise sum but axis=1 represents column
```

```python
# where in numpy
arr=np.array([10,-5,20,-3,15])
result=np.where(arr<0,0,arr)#(condition,true,else)
print(result)
```

```python
result=np.where(arr%2!=0,-1,arr)
print(result) 
```

```python
data=np.array([10,20,30,40,50])
mean_data=np.mean(data)
result=np.where(data>mean_data,mean_data,data)
print(result)

```
