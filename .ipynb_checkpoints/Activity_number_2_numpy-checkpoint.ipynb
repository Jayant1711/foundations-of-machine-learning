```python
import numpy as np

marks=np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
print(marks)
```

```python
# Task A1
print(marks.shape)
print(marks.size)
print(marks.dtype)

# task A2
#  marks o 2nd student

print(marks[1:2])
# marks o 3rd subject all students

print(marks[:,2:])
# student 1 subject 2 marks

print(marks[0:1,1:2])

```

```python
# task b1
# ag marks o each student and subject

print(marks.mean(axis=1))
print(marks.mean(axis=0))
```

```python
# task b2
d=np.where(((marks.mean(axis=1))>=80),marks,0)
print(d)
# error
```

```python
# task c1
status_matrix=np.where(marks>=50,1,0)
print(status_matrix)

# task c2

marks_cleaned=np.where(marks<40,0,marks)
print(marks_cleaned)
```
