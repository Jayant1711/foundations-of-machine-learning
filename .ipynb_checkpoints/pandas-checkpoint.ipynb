```python
import pandas as pd
import numpy as np
```

```python
s=pd.Series([10,20,30,40,50],index=["a","b","c","d","e"])
print(s)
print(s["c"])
print(s[2])
```

```python
new=np.array([10,3,48,39,93,39,59,29,49,28,42,56])
df=new.reshape(4,3)
a=pd.DataFrame(df,index=["a","s","d","r"],columns=["A","b","c"])
# a.index(["a","b","c","d"])
print(a)
```

```python
# series by dictionary

dict={"Math":25,"Phy":45,"che":96}

d=pd.Series(dict)
print(d)
```

```python
aa={
    "name":["a","b","c"],
    "Age":[10,20,30],
    "marks":[20,30,50]
}
x=pd.DataFrame(aa,index=["la","la","la"])
print(x)
```

```python
x.set_index("name", inplace=True)
# inplace == true means we are changing the original object

print(x)
```
