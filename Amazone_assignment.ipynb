```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,'r') as f:
        data=json.load(f)
    return data
```

```python
data=load_data("store_data.json")
print(data)
```

```python
# clean and strcture the data
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    for user in data:
        # clean strings
        raw_rating= str(user["rating"].strip().lower())
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"]=str(raw_rating)
        # rating string to number done
        # now what about missing age lets store none in that
        raw_age=user.get("age")
        if(raw_age== None):
            user["age"] = None    
         # now handling the duplicate data which can here be only feedback by same user 2nd time that to samme
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
            
        cleaned_data.append(user)
    return cleaned_data
        
        
        
```

```python
clean=clean_data(data)
print(clean)
 
```

```python
#to get clean insights
def get_insights(clean):
    total_rating=0
    for user in clean:

        # got the average rating for the product
        total_rating += float(user["rating"])
    print(f"avg rating={total_rating/len(clean)}")
        
 # poor ratings percentage given by users
    poor_rating=0
    for user in clean:
        if(float(user["rating"])<3):
            poor_rating+=1
    print(f"% of user with poor rating={poor_rating/len(clean)*100}%")
        
        
        
    
```

```python
get_insights(clean)
```

```python
# trying to make a product recommendation feature(rule based, not machine learning)

def get_recommendation(clean):
    
    reccomendation=[]

    for user in clean:
        curr_recom={}
        curr_recom["name"]=user["name"]
         
        if(float(user["rating"])>=4):
              curr_recom["brand"]="Apple"
        else:
            curr_recom["brand"]="Samsung"
        
        reccomendation.append(curr_recom)
    
    
    return reccomendation
        
```

```python
get_recommendation(clean)
```
